# Move all CSVs from duplicate "glacier_snow_cover_exports" folders into one folder, and then remove the duplicate folders

```python
import os
from glob import glob
from tqdm import tqdm
```

```python
# Mount your Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# View all the duplicate folders in your Drive
base_folder = '/content/drive/MyDrive'
folder_name = 'glacier_snow_cover_exports'
merged_folder = 'snow-mapping'
# get all the folders with "glacier_snow_cover_exports" in the folder name
folders = [x for x in os.listdir(base_folder) if folder_name in x]
print(f'Found {len(folders)} total folders')
# folders
```

```python
print(f'Base folder name: {os.path.join(base_folder, folder_name)}')
files = glob(os.path.join(base_folder, folder_name + '*', '*.csv'))
print(f'{len(files)} files in the base folder...')
```

```python
# Move all files into the main folder
files = glob(os.path.join(base_folder, folder_name + '*', '*.csv'))
print(f'Moving {len(files)} files into the main folder...')
for file in tqdm(files):
  os.rename(file, os.path.join(base_folder, merged_folder, os.path.basename(file)))
print('Done!')
```

```python
# When finished copying, delete the duplicate folders
# the duplicate folders have a " (1)", etc. at the end of the name
folders_remove = glob(os.path.join(base_folder, folder_name + ' *'))
print(f'Removing {len(folders_remove)} folders...')
for folder in tqdm(folders_remove):
  # print(folder)
  os.rmdir(folder)
print('Done!')
```
